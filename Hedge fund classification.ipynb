```python
# import libraries
import numpy as np
import pandas as pd
from NNDeep import DeepNN
import time
```

```python
df = pd.read_csv('deepanalytics_dataset.csv')
df.head()
```

```python
y = df['target']
X = df.drop(['data_id','period','target'],axis=1)
print('X', X.head())
print('y', y.head())
```

```python
X = np.array(X)
y = np.array(y).reshape(len(X),1)
```

```python
NN = DeepNN([88,100,200,100,1])
w = NN.learn(epochs=100, learning_rate=0.001, X=X, y=y, activations=['reLU','sigmoid','sigmoid','sigmoid'],\
             cost_func='log_loss', metrics_at=1, optimizer='rmsprop', batch_size=30,split=True)
```
